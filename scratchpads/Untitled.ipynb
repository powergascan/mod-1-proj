```python
#Let's not use rotten tomato for now
spy_monthly=pd.read_csv("SPY.csv")
```

```python
spy_monthly.head()
```

```python
spy_monthly['view_date']=pd.to_datetime(spy_monthly['Date'])
spy_yearly=spy_monthly[spy_monthly['view_date'].dt.month==1]
spy_yearly['year']=spy_yearly['view_date'].dt.year
```

```python
spy_monthly['percent_return_month']=(spy_monthly['Adj Close']-spy_monthly['Adj Close'].shift(1))/spy_monthly['Adj Close']*100
spy_yearly['percent_return_year']=(spy_yearly['Adj Close']-spy_yearly['Adj Close'].shift(1))/spy_yearly['Adj Close']*100
```

```python
matched=matched.merge(spy_monthly[['view_date','percent_return_month']], on="view_date")
matched=matched.merge(spy_yearly[['year','percent_return_year']], on="year")
```

```python
to_replace_list=["domestic_gross","worldwide_gross","production_budget"]
for i in to_replace_list:
    matched[i]=matched[i].str.replace(r"$","")
    matched[i]=matched[i].str.replace(r",","")
```

```python
GDP=pd.read_csv("../GDPDEF.csv").rename(columns={"Year":"year"})
matched=matched.merge(GDP, on='year')
```

```python
for i in to_replace_list:
    matched[i]=matched[i].astype("int64")/matched['GDP_Index']
```

```python
matched['domestic_profit']=(matched['domestic_gross']-matched['production_budget'])
matched['foreign_profit']=(matched['worldwide_gross']-matched['domestic_gross'])
matched['profit_margin']=(matched['worldwide_gross']-matched['production_budget'])/(matched['production_budget'])*100
```

```python
imdb_title_ratings.rename(columns={"tconst":"title_id"}, inplace=True)
imdb_title_basics.rename(columns={"tconst":"title_id"}, inplace=True)
imdb_title_principals.rename(columns={"tconst":"title_id"}, inplace=True)
matched_full=matched.merge(imdb_title_ratings, on='title_id', how='left')
matched_full=matched_full.merge(imdb_title_basics, on='title_id', how='left')
#matched_full=matched_full.merge(imdb_title_principals, on='title_id', how='left')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
matched['production_budget_mean']=matched.groupby(['year'])['production_budget'].transform("mean")
matched['production_budget_std']=matched.groupby(['year'])['production_budget'].transform("std")
matched['production_budget_norm']=(matched['production_budget']-matched['production_budget_mean'])/matched['production_budget_std']
matched['budget_size']=pd.cut(matched['production_budget_norm'], bins=3)
```

```python
plt.figure(figsize=(12,10))
ax=sns.boxplot(y="profit_margin",x="year", data=matched,showfliers=False, hue='budget_size')
ax2 = ax.twinx()
sns.barplot(y='percent_return_year',x="year",data=spy_yearly, color="black", ax=ax2, alpha=0.05)
plt.xticks(rotation=90)
```
