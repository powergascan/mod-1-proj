```python
tn_movie_budgets.rename(columns={"movie":"title"}, inplace=True)
matched=tn_movie_budgets
matched['release_date']=pd.to_datetime(matched['release_date'])
matched['view_date']=matched.release_date+pd.offsets.MonthBegin(1)
matched['year']=matched['view_date'].dt.year
matched=matched[matched['year']<=2018]
```

```python
spy_monthly['view_date']=pd.to_datetime(spy_monthly['Date'])
spy_monthly.sort_values(["view_date"],inplace=True)
spy_yearly=spy_monthly[spy_monthly['view_date'].dt.month==1]
spy_yearly['year']=spy_yearly['view_date'].dt.year
```

```python
spy_monthly['percent_return_month']=(spy_monthly['Adj Close'].shift(-1)-spy_monthly['Adj Close'])/spy_monthly['Adj Close']*100
spy_yearly['percent_return_year']=(spy_yearly['Adj Close'].shift(-1)-spy_yearly['Adj Close'])/spy_yearly['Adj Close']*100
spy_monthly['percent_return_rolling']=spy_monthly.percent_return_month.rolling(2).mean()
spy_monthly['market_condition']=pd.cut(spy_monthly['percent_return_rolling'], bins=3, labels=['Low','Medium','High'])
```

```python
matched=matched.merge(spy_monthly[['view_date','percent_return_month','percent_return_rolling',"market_condition"]], on="view_date")
matched=matched.merge(spy_yearly[['year','percent_return_year']], on="year")
```

```python
to_replace_list=["domestic_gross","worldwide_gross","production_budget"]
for i in to_replace_list:
    matched[i]=matched[i].str.replace(r"$","")
    matched[i]=matched[i].str.replace(r",","")
```

```python
GDP=pd.read_csv('..//download_disposable_income.csv')
matched=matched.merge(GDP, on='year')
```

```python
for i in to_replace_list:
    matched[i]=matched[i].astype("int64")/matched['nondurable_goods_expenditure']*100
```

```python
matched['domestic_profit']=(matched['domestic_gross']-matched['production_budget'])
matched['foreign_profit']=(matched['worldwide_gross']-matched['domestic_gross'])
matched['profit_margin']=(matched['worldwide_gross']-matched['production_budget'])/(matched['production_budget'])*100
matched['domestic_margin']=(matched['domestic_profit'])/(matched['production_budget'])*100
matched['foreign_margin']=(matched['foreign_profit'])/(matched['production_budget'])*100
```

```python
matched['production_budget_rank']=matched.groupby(['year'])['production_budget'].rank(pct=True)
matched['budget_size']=pd.cut(matched['production_budget_rank'], bins=2, labels=["Independent","MassProduction"])
matched['count']=matched.groupby(['view_date'])['title'].transform("count")
matched['competition_size']=pd.cut(matched['count'], bins=4, \
                                   labels=['Low','Moderately Low','Moderately High','High'])
```

```python
matched.hist('count',by="budget_size")
```

```python
plt.clf()
plt.figure(figsize=(18, 12))
sns.boxplot('competition_size','profit_margin', data=matched, hue="budget_size", showfliers=False)
plt.xlabel("Competition Level", size=20)
plt.ylabel("Profitability", size=20)
plt.title("Competition and Profitability for Independent vs Mass Production Movies", size=30)
plt.savefig("..\\competition and profitability.png")
```

```python
plt.clf()
plt.figure(figsize=(18, 12))
sns.boxplot('market_condition','domestic_margin', data=matched, hue="budget_size", showfliers=False)
plt.xlabel("Market Condition", size=20)
plt.ylabel("Profitability", size=20)
plt.title("Market Condition and Profitability for Independent vs Mass Production Movies", size=30)
plt.savefig("..\\Market and profitability.png")
```

```python
matched_agg=matched.groupby("view_date").agg({'profit_margin':"mean","id":"nunique"}).reset_index()
matched_agg['month']=matched_agg.view_date.dt.strftime("%B")
matched_agg['month_num']=matched_agg.view_date.dt.month
matched_agg['count']=matched_agg.groupby("month")['id'].transform(sum)
matched_agg.describe()
matched_agg_month=matched_agg.groupby("month").\
        agg({"count":"min","profit_margin":"mean","month_num":"min"}).reset_index()
matched_agg_month=matched_agg_month.sort_values("month_num")
```
